```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
class AudioAugmentation:
    
    def read_audio_file(self, file_path):
        data, sr = librosa.core.load(file_path)
        return data
    
    def stretch(self, data, rate=1):
        data = librosa.effects.time_stretch(data, rate)
        return data
    
    def pitch_shift(self, data, sr, n_steps):
        data = librosa.effects.pitch_shift(data, sr, n_steps)
        return data
    
    def write_audio_file(self, file, data):
        librosa.output.write_wav(file, data)
        
    def plot_time_series(self, data):
        fig = plt.figure(figsize=(14, 8))
        plt.title('Raw wave ')
        plt.ylabel('Amplitude')
        plt.plot(np.linspace(0, 1, len(data)), data)
        plt.show()
        
```

```python
aa = AudioAugmentation()

data = aa.read_audio_file("Разве это Наташа.wav")
aa.plot_time_series(data)
ipd.Audio(data, rate=22050)
```

```python
data_stretch = aa.stretch(data, 0.9)
aa.plot_time_series(data_stretch)
ipd.Audio(data_stretch, rate=22050)
```

```python
data_stretch = aa.stretch(data, 1.29)
aa.plot_time_series(data_stretch)
ipd.Audio(data_stretch, rate=22050)
```

```python
data_pitch = aa.pitch_shift(data, 22050,  n_steps=3)
aa.plot_time_series(data_pitch)
ipd.Audio(data_pitch, rate=22050)
```
